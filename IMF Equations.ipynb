Definition of the Chabrier mass function


Chabrier 2003 gives:

\begin{equation}
\xi(\log m) \equiv \frac{dN}{d\log M}= A \exp \left[-(\log m - \log m_c)^2 / 2\sigma^2 \right]
\end{equation}

the `imf` package is presently defining instead:
\begin{equation}
\xi(m) \equiv \frac{dN}{dM}= A \exp \left[-(\log m - \log m_c)^2 / 2\sigma^2 \right]
\end{equation}

This is equivalent to 

$$
\frac{M}{\ln 10} \frac{dN}{dM} = A \exp \left[-(\log m - \log m_c)^2 / 2\sigma^2 \right]
$$

so 
$$
 \frac{dN}{dM} = A \frac{\ln 10}{M}\exp \left[-(\log m - \log m_c)^2 / 2\sigma^2 \right]
$$

as always, $$\xi_{\log} = \frac{\ln 10}{M} \xi$$
